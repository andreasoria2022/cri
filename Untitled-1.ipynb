BITCOIN

Vamos a extraer de la API CoinGeckoAPI precios de la criptomoneda bitcoin por fecha 

```python
#Utilizare la biblioteca CoinGeckoAPI para extraer los precios hist√≥ricos en d√≥lares
cg = CoinGeckoAPI()

#Desde el 28 de diciembre de 2013 hasta el 19 de agosto de 2023
start_date = datetime.datetime(2013, 12, 27)
end_date = datetime.datetime(2023, 8, 19)

start_timestamp = int(start_date.timestamp())
end_timestamp = int(end_date.timestamp())

#Los precios se extraen utilizando la funci√≥n ‚Äúget_coin_market_chart_range_by_id‚Äù.
bitcoin_prices = cg.get_coin_market_chart_range_by_id(id='bitcoin', vs_currency='usd', from_timestamp=start_timestamp, to_timestamp=end_timestamp)

#Se escriben en un archivo CSV llamado ‚Äúbitcoin_prices.csv‚Äù
#La funci√≥n ‚Äúcsv.writer‚Äù se utiliza para escribir los datos en el archivo CSV
with open('bitcoin_prices.csv', mode='w') as bitcoin_file:
    bitcoin_writer = csv.writer(bitcoin_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    bitcoin_writer.writerow(['Date', 'Price'])

    for price in bitcoin_prices['prices']:
        bitcoin_writer.writerow([price[0], price[1]])
        print(price)
```

Vamos a extraer de la API CoinGecko la capitalizacion por fecha de la criptomoneda Polkadot

Utilizare la biblioteca requests para hacer una solicitud a la API de CoinGecko y extraer los datos de capitalizaci√≥n de mercado de polkadot en d√≥lares estadounidenses desde el 19 de agosto de 2020 hasta el 19 de agosto de 2023. Los datos se escriben en un archivo CSV llamado "polkadot_capitalizacion.csv‚Äù utilizando la funci√≥n ‚Äúcsv.writer‚Äù.

```python
import csv
import requests
from datetime import datetime

url = "https://api.coingecko.com/api/v3/coins/polkadot/market_chart/range"
params = {
    "vs_currency": "usd",
    "from": "1597719600",
    "to": "1692414000"
}

response = requests.get(url, params=params)
data = response.json()

market_caps = [item[1] for item in data["market_caps"]]
dates = [item[0] for item in data["market_caps"]]

with open("polkadot_capitalizacion.csv", mode="w") as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Market Cap"])
    for i in range(len(dates)):
        writer.writerow([dates[i], market_caps[i]])
```

Vamos a extraer de la API CoinGecko el volumen de negociacion de 24 horas de polkadot

Este c√≥digo utiliza la biblioteca requests para hacer una solicitud a la API de CoinGecko y extraer los datos de volumen de negociaci√≥n de tron en d√≥lares estadounidenses desde una fecha 19/08/2020 hasta la fecha 19/08/2023. Los datos se escriben en un archivo CSV llamado ‚Äúpolkadotvolumen_negociacion.csv‚Äù utilizando la funci√≥n ‚Äúcsv.writer‚Äù.

```python
import csv
import requests
from datetime import datetime

start_date = datetime(2020, 8, 17)
end_date = datetime(2023, 8, 19)

start_timestamp = int(start_date.timestamp())
end_timestamp = int(end_date.timestamp())

url = f'https://api.coingecko.com/api/v3/coins/polkadot/market_chart/range?vs_currency=usd&from={start_timestamp}&to={end_timestamp}'
response = requests.get(url)
data = response.json()

with open('polkadotvolumen_negociacion.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Date', '24 Hour Trading Vol'])
    for item in data['total_volumes']:
        writer.writerow([item[0], item[1]])
```
